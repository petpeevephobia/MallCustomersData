```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## LOAD DATA
path = 'Mall_Customers.csv'
df = pd.read_csv(path, index_col='CustomerID')
```

```python
## CLEAN DATA

# make column labels lower case
new_col = []
for col in df.columns:
    col = col.lower()
    new_col.append(col)

new_col[2] = 'annual_income'    
new_col[3] = 'spend_score'

df.columns = new_col
```

```python
sns.set_style("darkgrid")
df
```

# Regression Line Graph: Annual Income against Spend Score

Among males, spend_score does not increase as annnual_income increases. They do not spend more although they have higher annual income.

Among females, spend_score INCREASES as annual_income increases. They spend more with higher annual income.

```python
sns.lmplot(data=df, x='annual_income', y='spend_score', hue='gender')
```

# Categorical Scatter Plot: Annual Income of Males VS Females

Males generally have greater annual_income than females. Females have lesser annual_income.

```python
sns.swarmplot(data=df, x='gender', y='annual_income')
```

# Line Plot: Annual Income earned by Different Ages

annual_income is generally highest between the ages 35 and 45.

```python
sns.lineplot(x=df.age, y=df.annual_income)
```

# Regression Line Graph: Spend Score of Different Ages

The age range between 20 and 40 years spends MORE than the age range between 40 and 70 years.

The younger the customer's age, the more they spend.

```python
sns.regplot(data=df, x='age', y='spend_score')
```
